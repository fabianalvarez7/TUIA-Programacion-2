```python
class Stack:
    def __init__(self):
        self.items = []

    def push (self, item):
        self.items.append(item)

    def pop (self):
        return self.items.pop()

    def isEmpty (self):
        return (self.items == [])

    def __str__(self):
        return " ".join(str(i) for i in self.items)

    
```

```python
S = Stack()
S.push(11)
S.pop()
S.push(22)
S.push(33)
S.push(44)
S.pop()
print(S)
```
