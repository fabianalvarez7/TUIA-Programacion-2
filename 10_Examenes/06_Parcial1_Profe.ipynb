
# Ejercicio 1

Se necesita desarrollar un programa para gestionar una lista de estudiantes de una escuela. Se desea implementar una clase llamada Estudiante que tenga los métodos obtener_legajo y obtener_nombre para obtener el legajo y la edad del estudiante, respectivamente.

Además, se debe implementar una clase llamada Escuela con los siguientes métodos:

A. agregar_estudiante: recibe un objeto Estudiante y lo agrega a la lista de estudiantes de la escuela.

B. eliminar_estudiante: recibe el legajo de un estudiante y lo elimina de la lista de estudiantes de la escuela.

C. buscar_estudiante: recibe el legajo de un estudiante y devuelve True si el estudiante está en la lista de la escuela, False en caso contrario.

Nota Consideramos que el legajo es un identificador único dentro de cada escuela.

```python
l = [1, 2, 3]
for i, elem in enumerate(l):
  print(i, elem)
```

```python
class Escuela:
  def __init__(self) -> None:
    self.estudiantes = []

  def agregar_estudiante(self, estudiante: Estudiante) -> None:
    self.estudiantes.append(estudiante)

  def eliminar_estudiante(self, legajo: str) -> None:
    idx = 0
    for i, estudiante in enumerate(self.estudiantes):
      if estudiante.obtener_legajo() == legajo:
        idx = i

    self.estudiantes.pop(idx)

  def buscar_estudiante(self, legajo: str) -> bool:
    for estudiante in self.estudiantes:
      if estudiante.obtener_legajo() == legajo:
        return True

    return False

```

# Ejercicio 2

Implemente una función que reciba una lista enlazada L que contiene números enteros y un número entero k y devuelva una nueva lista enlazada en la que para cada posición i de dicha lista enlazada, contenga el resultado de la suma de i y los k − 1 elementos anteriores a i. En las primeras k − 1 posiciones, devolver la suma parcial hasta dicho momento.

La función debe ser pura, es decir, no debe modificar la lista recibida por parámetro.

Por ejemplo, si se llama a la función con la lista 1 −→ 2 −→ 3 −→ 4 −→ 5 y k = 3, el resultado debería ser 1 −→ 3 −→ 6 −→
9 −→ 12.

Nota: Puede considerar que están definidos todos los métodos de la interfaz de Lista (__str__, __len__, append,
insert, remove, pop e index).

**Primera idea**: Llevamos una lista auxiliar val_k que vaya llevando los valores que hay que sumar

```python
from typing import Any

class _Nodo:
  def __init__(self, dato: Any = None, prox: "_Nodo" = None) -> None:
    self.dato = dato
    self.prox = prox

class ListaEnlazada:
  def __init__(self) -> None:
    self.head = None
    self.len = 0

  def append(self, d: Any) -> None:
    nuevo_nodo = _Nodo(d)
    if self.head is None:
      self.head = nuevo_nodo
    else:
      nodo_actual = self.head
      while nodo_actual.prox is not None:
        nodo_actual = nodo_actual.prox
      nodo_actual.prox = nuevo_nodo
    self.len += 1

  def __str__(self):
    lista = ""
    nodo_actual = self.head
    while nodo_actual is not None:
      lista += str(nodo_actual.dato) + "->"
      nodo_actual = nodo_actual.prox
    return lista
```

```python
def f2(l: ListaEnlazada, k: int) -> ListaEnlazada:
  nueva_lista = ListaEnlazada()

  nodo_actual = l.head
  val_k = []
  i = 0

  while nodo_actual is not None:
    if i >= k:
      val_k.pop(0)

    val_k.append(nodo_actual.dato)
    suma_k = 0
    for j in val_k:
      suma_k += j
    nueva_lista.append(suma_k)
    nodo_actual = nodo_actual.prox
    i += 1

  return nueva_lista


L = ListaEnlazada()
L.append(1)
L.append(2)
L.append(3)
L.append(4)
L.append(5)
print(f2(L, 3))
```

**Complejidad de la primera idea**: Por cada elemento de la lista realizamos k sumas. Eso implica que si la lista tiene n elementos estamos realizando O(nk) operaciones

**Segunda idea** Una solución mejor puede encontrarse notando que para cada posición $i$, el valor que debemos escribir es:

$$
\sum_{j=i-k +1}^i L[j]
$$

y esto es igual a:
$$
\sum_{j=1}^i L[j] - \sum_{j=1}^{i-k} L[j]
$$

Podemos aprovechar esto y llevar dos sumas parciales, y calcular el valor en cada posición a partir de ellas. Esto ocasiona que solo realicemos como mucho dos sumas en cada paso (independientemente de k) y asi conseguimos una complejidad temporal O(n)

```python
def f2a(l: ListaEnlazada, k: int) -> ListaEnlazada:
  nueva_lista = ListaEnlazada()
  suma_k = 0
  resta_k = 0

  nodo_actual_resta = l.prim
  nodo_actual_suma = l.prim
  i = 0
  while nodo_actual_suma is not None:
    suma_k += nodo_actual_suma.dato
    nodo_actual_suma = nodo_actual_suma.prox

    if i > k:
      resta_k += nodo_actual_resta.dato
      nodo_actual_resta = nodo_actual_resta.prox
    i += 1

    valor = suma_k - resta_k
    nueva_lista.append(valor)

  return nueva_lista

L = ListaEnlazada()
L.append(1)
L.append(2)
L.append(3)
L.append(4)
L.append(5)
print(f2(L, 3))
```

# Ejercicio 3

Implemente una función que reciba un árbol binario A y que dado un valor entero M, una clave inicial inicio y una clave final fin, se devuelva una lista con todos los datos cuyas claves estén entre inicio y fin, que estén dentro de los primeros M niveles del árbol (considerando a la raíz en nivel 0).

Por ejemplo, dado el siguiente árbol, inicio = 5 , fin = 15 y M = 2, debería dar la lista [6, 8, 10, 14] (1 y 3 quedan afuera porque son menores que inicio, 4 y 7 quedan afuera porque están en el nivel 3).

```python
    8
   3 10
  1 6 14
 4 7
```

**Ayuda** Defina la función de manera recursiva e incorpore un parámetro para control del nivel actual.

```python
class BTree:
  def __init__(self, dato=None, left=None, right=None):
    self.dato = dato
    self.right = right
    self.left = left

def f3(A: BTree, M: int, inicio: float, fin: float, nivel = 0) -> list[int]:
  if A is None:
    return []

  if nivel > M:
    return []

  lista = []

  nivel +=1

  lista_izq = f3(A.left, M, inicio, fin, nivel)
  lista_der = f3(A.right, M, inicio, fin, nivel)
  if inicio <= A.dato <= fin:
    lista.append(A.dato)

  return lista_izq + lista + lista_der

tree = BTree(8, BTree(3, BTree(1), BTree(6, BTree(4), BTree(7))), BTree(10, BTree(6), BTree(14)))
print(f3(tree, 2, 5, 15))

```
