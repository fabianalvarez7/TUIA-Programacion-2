Implemente los siguiente métodos:

 *Ayuda: pensar que cada árbol tiene a su izquierda y derecha objetos árboles como sus hijos*.
* **nodos**: devuelve la cantidad de nodos del árbol
* **menor_mayor**: devuelve el menor y el mayor elemento del árbol en una tupla
* **buscar**: busca si un elemento está o no en el árbol
* **altura**: calcula la altura del árbol, la distancia desde la raíz hasta la hoja más lejana

````python
class Tree:
    def __init__(self, cargo, left=None, right=None):
        self.cargo = cargo
        self.left = left
        self.right = right

  #def __str__(self):
  #    return str(self.cargo)


    def nodos(self):
        cantidad = 1
        if self.left is not None:
            cantidad += self.left.nodos()

        if self.right is not None:
            cantidad += self.right.nodos()

        return cantidad


    def nodos2(self):
        if self.left is None and self.right is None:
            return 1

        elif self.left is None:
            return 1 + self.right.nodos2()

        elif self.right is None:
            return 1 + self.left.nodos2()

        else:
            return 1 + self.left.nodos2() + self.right.nodos2()


    def menor(self):
        minimo = self.cargo

        if self.left is not None:
            lminimo = self.left.menor()
            minimo = min(minimo, lminimo)

        if self.right is not None:
            rminimo = self.right.menor()
            minimo = min(minimo, rminimo)

        return minimo

    def mayor(self):
        maximo = self.cargo

        if self.left is not None:
            lmaximo = self.left.mayor()
            maximo = max(maximo, lmaximo)

        if self.right is not None:
            rmaximo = self.right.mayor()
            maximo = max(maximo, rmaximo)

        return maximo


    def menor_mayor(self):
        if self.left is None and self.right is None:
            return (self.cargo, self.cargo)

        elif self.left is None:
            rmin, rmax = self.right.menor_mayor()
            return (min(self.cargo, rmin), max(self.cargo, rmax))

        elif self.right is None:
            lmin, lmax = self.left.menor_mayor()
            return (min(self.cargo, lmin), max(self.cargo, lmax))

        else:
            lmin, lmax = self.left.menor_mayor()
            rmin, rmax = self.right.menor_mayor()
            return (min(self.cargo, lmin, rmin), max(self.cargo, lmax, rmax))


    # def buscar(self, element):
    #     if element == self.cargo:
    #         return True

    #     elif element < self.cargo:
    #         if self.left is None:
    #             return False
    #         else:
    #             return self.left.buscar(element)

    #     else:
    #         if self.right is None:
    #             return False
    #         else:
    #             return self.right.buscar(element)


    def buscar(self, element):
        if self.cargo == element:
            return True

        if self.left is not None:
            if self.left.buscar(element):
                return True

        if self.right is not None:
            if self.right.buscar(element):
                return True

        return False



# OTRA VERSION DE BUSCAR:
    #def buscar2(self, element):
    #     if self.cargo == element:
    #         return True
    #     elif self.left is not None and self.left.buscar2(element):
    #         return True
    #     elif self.right is not None and self.right.buscar2(element):
    #         return True
    #     else:
    #         return False



    def altura(self):
        if self.left is None and self.right is None:
            return 0

        elif self.left is None:
            return 1 + self.right.altura()

        elif self.right is None:
            return 1 + self.left.altura()

        else:
            return 1 + max(self.left.altura(),self.right.altura())



#        ___8___
#       /       \
#      2         15
#    /  \       /  \
#   4   -3    17    20
#  / \             /  \
# 7   3          22    0
tree = Tree(8, Tree(2, Tree(4, Tree(7), Tree(3)), Tree(-3)), Tree(15, Tree(17), Tree(20,Tree(22),Tree(0))))
````
