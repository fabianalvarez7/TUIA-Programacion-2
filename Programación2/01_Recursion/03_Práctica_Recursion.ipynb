**Ejercicio 5**

Convierta la siguiente función recursiva a una iterativa.

````python
def recursiva(t: int, k: int) -> int:
  if t >= 100:
    return k
  else:
    return recursiva(t + k, k + 1)

recursiva(3, 5)
````

````python
def iterativa(t, k):
    while t < 100:
        t += k
        k += 1
    return k

iterativa(3, 5)
````

**Ejercicio 6**

Escribir una función recursiva que calcule recursivamente el n-ésimo número triangular (el número 1 + 2 + 3 + ⋯ + n).

````python
def triangular (n: int) -> int:
    # Caso base: si n es 1, devolvemos 1 (el primer número triangular)
    if n == 1:
        return 1
    # Caso recursivo: llamamos a la función con n-1 y sumamos n al resultado
    else:
        return n + triangular(n-1)

triangular(20)
````

````python
from math import sqrt

def is_triangular(x):
    n = int((sqrt(8 * x + 1) - 1) / 2)
    return n * (n + 1) / 2 == x

def triangular(n):
    if n == 1:
        return 1
    else:
        result = n + triangular(n - 1)
        if is_triangular(result):
            print(f"{result} es un número triangular.")
        else:
            print(f"{result} no es un número triangular.")
        return result

is_triangular(10)
````

**Ejercicio 7**

Escribir una función recursiva que reciba un número positivo n y devuelva la cantidad de dígitos que tiene.

````python
def cantidad_digitos(n):
    # Caso base: si n es menor que 10, tiene un solo dígito
    if n < 10:
        return 1
    # Caso recursivo: llamamos a la función con el cociente de n entre 10 y sumamos 1 al resultado
    else:
        return 1 + cantidad_digitos(n // 10)

cantidad_digitos(12345)
````

Como 123 no es menor que 10, la función se llama a sí misma con cantidad_digitos(12).

Como 1 es menor que 10, la función devuelve 1.

**Ejercicio 8**

Convierta la siguiente función iterativa a una recursiva.

````python
def iterativa(l: list[int]) -> int:
    c = 1
    for i in l:
        c = c * i
    return c

iterativa([1,2,3,4,5,6,7,8,9])
````

````python
def recursiva(l: list[int]) -> int:
    if len(l) == 0:     # Caso base: si la lista está vacía, el producto es 0
        return 1   # por convención
    else:   # Caso recursivo: el producto es el primer elemento de la lista multiplicado por el producto del resto de la lista
        return l[0] * recursiva(l[1:])

recursiva([1,2,3,4,5,6,7,8,9])
````

La función recursiva(l) calcula el producto de todos los elementos de una lista de números enteros. La función tiene dos partes: el caso base y el caso recursivo. El caso base es una condición que se cumple cuando la lista está vacía. En este caso, la función devuelve 1, ya que por convención el producto de una lista vacía es 1.

En el caso recursivo, la función devuelve el primer elemento de la lista multiplicado por el producto del resto de la lista. Para obtener el producto del resto de la lista, se llama a la función recursiva con una porción de la lista que excluye al primer elemento.

Por ejemplo, si llamamos a la función con recursiva([1, 2, 3, 4]), la función se llamará a sí misma con recursiva([2, 3, 4]), luego con recursiva([3, 4]), y así sucesivamente hasta que se cumpla la condición del caso base. Finalmente, la función devolverá el producto de todos los elementos de la lista original.

**Ejercicio 9** Escribir una funcion recursiva que encuentre el mayor elemento de una lista.

````python
def encontrar_mayor(lista) -> int:
    if len(lista) == 0:
        return "La lista está vacía"
    elif len(lista) == 1:   # Caso base: Si la lista tiene un solo elemento, ese es el mayor
        return lista[0]
    else:
        mayor_sublista = encontrar_mayor(lista[1:])  # Llamada recursiva para encontrar el mayor en el resto de la lista
        if lista[0] > mayor_sublista:   # Compara el primer elemento con el mayor del resto de la lista
            return lista[0]
        else:
            return mayor_sublista


mi_lista = [4,9,2,7,5]
resultado = encontrar_mayor(mi_lista)
print("El mayor elemento de la lista es:", resultado)

````

La función se llama con la lista [4, 9, 2, 7, 5].

La función verifica la longitud de la lista. Dado que la lista tiene más de un elemento, no es un caso base, por lo que continúa con la recursión.

La función hace una llamada recursiva con la sublista [9, 2, 7, 5] (excluyendo el primer elemento, que es 4).

La nueva llamada recursiva trabaja con la sublista [9, 2, 7, 5]. Nuevamente, verifica la longitud de la lista y encuentra que no es un caso base.

Se realiza una segunda llamada recursiva con la sublista [2, 7, 5], excluyendo el primer elemento que es 9.

La tercera llamada recursiva trabaja con la sublista [2, 7, 5]. Nuevamente, no es un caso base.

Se realiza una cuarta llamada recursiva con la sublista [7, 5], excluyendo el primer elemento que es 2.

La quinta llamada recursiva trabaja con la sublista [7, 5]. Una vez más, no es un caso base.

Se realiza una sexta llamada recursiva con la sublista [5], excluyendo el primer elemento que es 7.

La sexta llamada recursiva encuentra que la longitud de la lista es igual a 1, por lo que es un caso base. En este caso, se devuelve el único elemento de la lista, que es 5.

La quinta llamada recursiva toma ese valor de 5 y lo compara con el primer elemento que excluyó, que es 7. Dado que 7 es mayor que 5, se devuelve 7 como el "mayor" de esta sublista.

La cuarta llamada recursiva toma el valor de 7 y lo compara con el primer elemento excluido, que es 2. Se devuelve 7 nuevamente.

La tercera llamada recursiva toma el valor de 7 y lo compara con el primer elemento excluido, que es 9. Esta vez, se devuelve 9, ya que 9 es mayor que 7.

La segunda llamada recursiva toma el valor de 9 y lo compara con el primer elemento excluido, que es 4. Se devuelve 9.

La primera llamada recursiva toma el valor de 9 y lo compara con el primer elemento de la lista original, que es 4. Nuevamente, se devuelve 9, ya que 9 es mayor que 4.

La llamada original a encontrar_mayor recibe el valor de 9 y lo compara con el primer elemento de la lista original, que es 4. Se devuelve 9 como el resultado final.

El resultado final de la función encontrar_mayor([4, 9, 2, 7, 5]) es 9, que es el mayor elemento de la lista. La recursión se usa para comparar y encontrar el elemento más grande entre todos los elementos de la lista.

## Recursión con wrappers

Al diseñar funciones recursivas muchas veces puede ser útil implementar una función **wrapper**, por ejemplo para cambiar los valores que hay que tomar, validar parámetros, inicializar datos o manejar excepciones.

Se utiliza una función recursiva, llamada funcion auxiliar, como base y luego una funcion no recursiva, llamada wrapper, que utiliza la funcion auxiliar para resolver lo pedido.

Por ejemplo, si necesitamos escribir una función recursiva para calcular el promedio, puede no ser evidente como hacerlo:

```python
def promediar(lista: list[float]) -> float:
  if len(lista) == 0:
    return ???
  promediar(lista[1:]) ???
```

Para calcular un promedio necesitamos computar tanto una suma como contar la cantidad de elementos. Podemos hacerlo con una función auxiliar

```python
def _promediar(lista: list[float]) -> int, float:
  if len(lista) == 0:
    return 0, 0
  cant, suma = promediar(lista[1:])
  return cant +1, suma + lista[0]

def promediar(lista: list[float]) -> float:
  cant, suma = _promediar(lista)
  return suma / cant
```
