```python
def hash_horner(clave, M=487): # M es el tama침o de la tabla - deberia ser primo!
        A = 37 # Este valor se puede cambiar, algunos valores elegidos usualmente son 37 y 256
        resultado = 0
        for caracter in clave:
            resultado = ord(caracter) + A * resultado
        return resultado % M


class Par:
    def __init__(self, clave, valor):
        self.clave = clave
        self.valor = valor

def iterador(lista, funcion):
        for elemento in lista:
            if funcion(elemento):
                return elemento
            return None

        

class TablaHash:
    def __init__(self, capacidad):
        self.capacidad = capacidad
        self.tabla = [[] for i in range (self.capacidad)]


    def __str__(self):
        texto = '{\n'
        for lista in self.tabla:
            for elemento in lista:
                texto += ' ' + str(elemento.clave) + ': ' + str(elemento.valor) + '\n'
        texto += '}'
        return texto


    def insertar(self, clave, valor):
        posicion = hash_horner(clave)
        # for par in self.tabla[posicion]:
        #     if par.clave == clave:
        #         par.valor = valor
        #         return
        # self.tabla[posicion].append (Par(clave, valor))
        lista = self.tabla[posicion]
        par_existente = iterador(lista, lambda x: x.clave == clave)
        if par_existente:
            par_existente.valor = valor
        else:
            lista.append(Par(clave, valor))


    def buscar(self, clave):
        posicion = hash_horner(clave)
        lista = self.tabla[posicion]
        par_existente = iterador(lista, lambda x: x.clave == clave)
        return par_existente.valor if par_existente else print('El producto no fue encontrado')


    def eliminar(self, clave):
        posicion = hash_horner(clave)
        del self.tabla[posicion]
        

    def estadisticas(self):
        pass



    # def menu() -> None:
    #     print("1. Add new item")
    #     print("2. Query item")
    #     print("3. Delete item")
    #     print("4. Stats")
    #     print("5. Exit")
    #     print()

    # if __name__ == "__main__":
    #     while True:
    #         menu()
    #         choice = int(input("Enter choice: "))
    #         if choice == 1:
    #             pass
    #         elif choice == 2:
    #             pass
    #         elif choice == 3:
    #             pass
    #         elif choice == 4:
    #             pass
    #         elif choice == 5:
    #             break
    #         else:
    #             print("Invalid choice")
    #         print()

```
