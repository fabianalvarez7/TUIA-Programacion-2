**Ejercicio 1** Implemente un algoritmo Divide y Vencerás que calcule la suma de los números de una lista.

```python
def es_caso_base(l) -> bool:
    # los casos base serán listas vacías o con un solo elemento
    return len(l) == 0 or len(l) == 1

def resolver_caso_base(l: list[int]) -> int:
    # devuelve 0 si la lista está vacía o el valor del único elemento
    if len(l) == 0:
        return 0
    return l[0]

def dividir(l: list[int]) -> tuple:
    # divide la lista en dos sublistas
    m = len(l) // 2
    sp1=l[:m]
    sp2=l[m:]
    return (sp1, sp2)

def combinar(s1: int, s2: int) -> int:
    return s1 + s2

def resolver(l: list[int])-> int:
    """
    Calcula la suma de los números de una lista mediante
    un algoritmo D&C.
    """
    if es_caso_base(l):
        return resolver_caso_base(l)
    else:
        sp1, sp2 = dividir(l)
        sol1 = resolver(sp1)
        sol2 = resolver(sp2)
        return combinar(sol1, sol2)
```

```python
num = [1, 2, 3, 4, 5, 10]

resolver(num)
```

**Ejercicio 2**  Implemente un algoritmo Divide y Vencerás que determine si un número $n$ (fijo) se encuentra o no en una lista NO ordenada de números.

```python
def es_caso_base(l: list[int]) -> bool:
    return len(l) == 0 or len(l) == 1

def resolver_caso_base(n: int, l: list[int]) -> bool:
    if len(l) == 0:
        return False
    if len(l) == 1:
        return l[0] == n

def dividir(l: list[int]) -> tuple[list]:
    m = len(l) // 2
    sp1=l[:m]
    sp2=l[m:]
    return sp1, sp2

def combinar(sol1: bool, sol2: bool) -> bool:
    return sol1 or sol2

def resolver(n: int, l: list[int]) -> bool:
    """
    Determina si n se encuentra en la lista.
    """
    if es_caso_base(l):
        return resolver_caso_base(n, l)
    else:
        sp1, sp2 = dividir(l)
        sol1 = resolver(n, sp1)
        sol2 = resolver(n, sp2)
        return combinar(sol1, sol2)
```

**Ejercicio 6**: Ordenamiento

Ordene una lista usando divide y vencerás, para esto preste atención cómo combina dos listas ordenadas para producir una nueva lista ordenada. No utilice ningún método de ordenamiento _builtin_.

```python
def es_caso_base(l: list[int]) -> bool:
    return len(l) <= 1

def resolver_caso_base(l: list[int]) -> list[int]:
    # si la lista tiene 0 o 1 elemento, ya está ordenada
    return l

def dividir(l: list[int]) -> tuple[list[int], list[int]]:
    m = len(l) // 2
    s1 = l[:m]
    s2 = l[m:]
    return s1, s2

def combinar(s1: list[int], s2: list[int]) -> list[int]:
    resultado = []
    i = 0
    j = 0
    while i < len(s1) and j < len(s2):
        if s1[i] < s2[j]:
            resultado.append(s1[i])
            i += 1
        else:
            resultado.append(s2[j])
            j += 1
    # agregar los elementos restantes de s1 o s2
    resultado.extend(s1[i:])
    resultado.extend(s2[j:])
    return resultado

def resolver(l: list[int]) -> list[int]:
    if es_caso_base(l):
        return resolver_caso_base(l)
    else:
        sp1, sp2 = dividir(l)
        s1 = resolver(sp1)
        s2 = resolver(sp2)
        return combinar(s1, s2)
```

```python
l = [4, 7, 89, 31, 55, 2, 107]
resultado = resolver(l)
print(resultado)
```

**Ejercicio 7**: Sublista de suma máxima.

Dada una lista de números, el problema de encontrar la sublista consiste encontrar una sublista contigua de forma tal que la suma de sus elementos sea la máxima posible.

Por ejemplo, para la lista [−2, 1, −3, 4, −1, 2, 1, −5, 4], la sublista de suma máxima es: [4, −1, 2, 1], que suma 6. Puede intentar convencerse de esto tomando otras posibles sublistas y notando que la suma siempre es menor o igual.

Formalmente, dada una lista $L$ de números enteros, el problema trata sobre encontrar índices $i, j$ con $0 \le i \le j \le n$ de forma tal que la suma

$$
\sum_{x=i}^j L[x]
$$

sea lo mas grande posible.

El algoritmo trivial para resolver este problema consiste en recorrer todos los posibles índices $i, j$, calcular la suma correspondiente en cada caso y elegir la máxima. Este algoritmo tiene complejidad temporal cúbica ( $O(n^3)$ ), es decir la cantidad de sumas que hay que resolver crece como una función cúbica del tamaño de la lista de entrada.

Sin embargo, utilizando Divide y Vencerás puede resolver este problema con complejidad temporal O(n log(n)) (llamada complejidad subcuadratica).  

Para poder lograrlo, necesitará en cada paso, más información que solo la solución final. La llamada recursiva debe devolver, si representamos una sublista como $L[i:j]$:
- La suma máxima $L[i:j]$ (la "solución final").
- La suma total $L[0:n]$
- La suma máxima por izquierda $L[0:j']$
- La suma máxima por derecha $L[i':n]$


Implemente la solución divide y vencerás siguiendo estas ideas.

```python
# definición del tipo de Solución
Solucion = tuple[int, int, int, int]


def es_caso_base(problema: list[int]) -> bool:
    return len(problema) == 1


def resolver_caso_base(problema: list[int]) -> Solucion:
    # devuelve una tupla con la suma maxima, la suma total,
    # la suma maxima por izquierda y la suma maxima por derecha
    elemento = problema[0]
    return (elemento, elemento, elemento, elemento)


def dividir(problema: list[int]) -> tuple[list[int], list[int]]:
    m = len(problema) // 2
    return problema[:m], problema[m:]


def combinar(s1: Solucion, s2: Solucion) -> Solucion:
    """
    Calcula la suma total, la suma maxima por izquierda,
    la suma maxima por derecha y la suma maxima de la lista combinada.
    """

    suma_maxima1, suma_total1, suma_max_izq1, suma_max_der1 = s1
    suma_maxima2, suma_total2, suma_max_izq2, suma_max_der2 = s2

    suma_total = suma_total1 + suma_total2
    suma_max_izq = max(suma_max_izq1, suma_total1 + suma_max_izq2)
    suma_max_der = max(suma_max_der2, suma_total2 + suma_max_der1)
    suma_maxima = max(suma_maxima1, suma_maxima2, suma_max_der1 + suma_max_izq2)

    return (suma_maxima, suma_total, suma_max_izq, suma_max_der)


def resolver(problema: list[int]) -> Solucion:
    """
    Divide la lista, resulve los subproblemas y combina las soluciones.
    """

    if es_caso_base(problema):
        return resolver_caso_base(problema)
    else:
        subproblema1, subproblema2 = dividir(problema)
        solucion1 = resolver(subproblema1)
        solucion2 = resolver(subproblema2)
        return combinar(solucion1, solucion2)
```

```python
lista = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
suma_maxima, suma_total, suma_max_izq, suma_max_der = resolver(lista)
print(f"La suma maxima es: {suma_maxima}")
```

**Ejercicio 8** Escriba un algoritmo Divide y Vencerás que tome como entrada una matriz de números (representada en Python como una lista de listas), y un número $n$ y determine cuantas veces aparece el número $n$ en la matriz.

**Ejercicio de parcial**

El número combinatorio C(n,k) se puede definir recursivamente como 1 si k = 0 o k = n, y C(n−1,k−1)+C(n−1,k) para cualquier otro valor de k.

a. Escriba un algoritmo de estilo Divide y Vencerás para calcular el número combinatorio C(n,k). Anotar el tipo de los parámetros y de retorno en todas las funciones.

b. Analice informalmente la eficiencia del algoritmo desarrollado.

```python
def es_caso_base(n: int, k: int) -> bool:
    return k == 0 or k == n

def resolver_caso_base(n:int, k: int) -> int:
    return 1

def dividir(n:int, k: int) -> tuple[tuple[int, int], tuple[int, int]]:
    return (n-1, k-1), (n-1, k)

def combinar(subproblema1: int, subproblema2: int) -> int:
    return subproblema1 + subproblema2

def resolver(n: int, k: int) -> int:
    if es_caso_base(n, k):
        return resolver_caso_base(n, k)

    subproblema1, subproblema2 = dividir(n, k)
    solucion1 = resolver(*subproblema1)
    solucion2 = resolver(*subproblema2)
    return combinar(solucion1, solucion2)
```

```python
n = 10
k = 4

print(f"C({n}, {k}) = {resolver(n, k)}")
```

```python
# Version simplificada

def combinatorio(n: int, k:int) -> int:
    # caso base
    if k == 0 or k == n:
        return 1

    # caso recursivo
    return combinatorio(n-1, k-1) + combinatorio(n-1, k)
```

```python
n = 10
k = 4

print(f"C({n}, {k}) = {combinatorio(n, k)}")
```

```python
# Casos de prueba de ambas funciones
for n in range(6):
    for k in range(n + 1):
        print(f"C({n}, {k}) = {resolver(n, k)} == {combinatorio(n, k)}")
```
