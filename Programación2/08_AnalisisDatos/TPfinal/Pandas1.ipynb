### Paquete CSV

Comencemos leyendo el archivo con csv. Como podr√°n recordar, extraer informaci√≥n puede ser un poco tedioso ya que para poder obtener todos los valores en una columna, tenemos que recorrer todas las filas del archivo csv.

En este caso vamos a recolectar los precios y los precios clasificados por barrios

```python
import csv
lista_precios = []
precio_por_barrio = {}

with open('listings.csv') as File:
    next(File)
    reader = csv.reader(File, delimiter=',')
    for fila in reader:
        precio = float(fila[9].replace('$', '').replace(',',''))
        lista_precios.append(precio)
        #por barrio
        barrio = fila[5] # neighbourhood
        if barrio in precio_por_barrio:
            precio_por_barrio[barrio].append(precio)
        else:
            precio_por_barrio[barrio] = [precio]
```

**En** el c√≥digo de arriba generamos una lista con todos los precios que aparecen en el archivo listing.csv y un diccionario donde las claves son los barrios y los valores son todos los precios de ese barrio. A continuaci√≥n hacemos una breve exploraci√≥n de esos datos

```python
import matplotlib.pyplot as plt
plt.hist(lista_precios)
```

```python
sum(lista_precios)/len(lista_precios)
```

```python
precio_por_barrio.keys()
```

```python
# Calculamos el precio promedio en el barrio Bijlmer-Centrum
sum(precio_por_barrio['Bijlmer-Centrum'])/len(precio_por_barrio['Bijlmer-Centrum'])
```

## La mutabilidad

En la unidad de Programaci√≥n Orientada a Objetos, vimos que exist√≠an dos tipos de m√©todos:

* Los **m√©todos puros** que no modifican el estado del objeto sobre el cu√°l lo llamamos.
* Los **m√©todos modificadores** que modifican el estado del objeto sobre el cu√°l lo llamamos.

Por si est√°n olvidados, veamos un ejemplo con una lista de python. Para ordenar una lista, podemos utilizar el m√©todo `.sort` ...

```python
A = [1, 2, 3, 4, 2, 1, 1]
A.sort()
print(A)
```

... que modifica la lista "in place". En el mismo objeto A se reordenan los elementos. O bien podemos hacer:

```python
A = [1, 2, 3, 4, 2, 1, 1]
A_ordenada = sorted(A)
print(A_ordenada)
```

... que _devuelve_ o _retorna_ una nueva lista con los mismos elementos que A, pero en orden. Notar que aca la lista original A **sigue existiendo**

```python
print(A)
```

Si queremos, podemos "pisar" la lista A con los valores ordenados simplemente asignando la misma variable al resultado

```python
A = [1, 2, 3, 4, 2, 1, 1]
A = sorted(A)
print(A)
```

En pandas ocurre lo mismo. Casi todos los m√©todos que modifiquen un dataframe o una serie aceptar√°n una versi√≥n `inplace` pasando el argumento `inplace=True` al m√©todo. Si no lo hacemos, pandas se comportar√° _retornando_ un nuevo objeto con las modificaciones que le pedimos. Luego ser√° nuestra responsabilidad asignar ese objeto a una variable (que puede o no ser la misma, dependiendo de lo que querramos).

```python
# data.set_index('id', inplace = True) #cambia el √≠ndice inplace
data = data.set_index("id")
data.head(2)
```

**Convenci√≥n**: Por convenci√≥n, evitaremos el uso de `inplace=True` y trabajaremos con m√©todos puros y objetos nuevos siempre que sea posible.

## Indexar
Pandas nos permite acceder a la informaci√≥n dentro de sus DFs y Series mediante sus √≠ndices con los atributos .loc y .iloc. `loc` nos permite acceder a los √≠ndices por su valor y `iloc` por su posici√≥n.
Vamos a poder acceder a un √≠ndice o a varios. Y tambi√©n vamos a poder incluir las columnas

* Primer sabor de loc: Si le pasamos un √≠ndice con un valor particular nos devuelve una Serie, donde el √≠ndice de la serie son las columnas del dataframe.

```python
data.loc[31553121]
```

* Segundo sabor de loc: Podemos pasarle una lista de √≠ndices y nos devuelve un dataframe solo con aquellos √≠ndices

```python
data.loc[[23726706, 35815036]]
```

* Tercer sabor de loc: Podemos pasarle, ademas, qu√© columna nos interesa. En este caso nos devolver√° una serie solo con los √≠ndices que le pedimos.

```python
data.loc[[31553121], 'price']
```

* Cuarto sabor de loc: Podemos pasarle una lista de las columnas que queremos. En este caso, nos devolvera un dataframe, que contiene solo los √≠ndices que le pedimos y solo las columnas que le pedimos.

```python
data.loc[[23726706, 35815036], ['price']]
```

**Convenci√≥n** Si bien `iloc`, que sirve para acceder a los √≠ndices y a las columnas por su posici√≥n y no por su nombre, puede ser de utilidad en ciertas situaciones, _lo evitaremos_ siempre que sea posible.

# Filtrado de datos

Muchas veces no queremos realizar el an√°lisis sobre todo el dataFrame, sino solo sobre una porci√≥n. Para ello Pandas nos permite realizar filtrados.

El filtrado se hace por condiciones, es decir nos quedamos con el subconjunto de datos que cumple cierta restricci√≥n. Por ejemplo, para obtener todos los departamentos que tiene al menos dos reviews por mes:

```python
data[ data['reviews_per_month'] >= 2 ]
```

Se puede filtar por condiciones **compuestas**, es decir por proposiciones complejas que involucren operadores l√≥gicos. Los operadores l√≥gicos son los siguientes:

* Y: se deben cumplir ambas condiciones para que el dato est√© en el resultado, similar al `and` que hacemos en los condicionales de Python, pero ac√° se utiliza el operador ampersand '&'.
* O: se debe cumplir al menos una de las dos condiciones para que el dato est√© en el resultado, similar al `or` que hacemos en los condicionales de Python, pero ac√° se utiliza el operador pipe '|' .
* NO: no se debe cumplir la condicion para que el dato est√© en el resultado, similar al `not` nativo, solo que se utiliza la virgulilla como operador '~'.

Por ejemplo, podemos obtener todos los departamentos que tengan al menos 2 reviews por mes y adem√°s al menos 90 reviews en total de la siguiente forma:

```python
data[ (data['reviews_per_month'] >= 2) & (data['number_of_reviews'] >= 90) ]
```

**Nota**: Las condiciones para filtrar se pueden complejizar tanto como se desee. Es importante encerrar entre par√©ntesis cada condici√≥n para evitar errores.

Para filtrar por valores num√©ricos, podemos utilizar los operadores de comparaci√≥n. Para filtrar por strings, tambi√©n podemos filtrar por igualdad. Por ejemplo, podemos obtener todos los departamentos que tengan `room_type` igual a "Private room":

```python
data[ data['room_type'] == "Private room" ]
```
