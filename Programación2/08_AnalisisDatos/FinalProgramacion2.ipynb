```python
# versión recursiva
def f(n: int, d: int) -> None:
    if n > 1:
        if n % d == 0:
            print(d)
            f(n//d, d)
        else:
            f(n, d+1)
    return

f(30, 2)
```

```python
# versión iterativa
def f(n: int, d: int) -> None:
    """
    Calcula los factores primos de n.
    """
    while n > 1:
        if n % d == 0:
            print(d)
            n //= d
        else:
            d += 1

f(30, 2)
```

```python
class BinaryTree:
    def __init__(self, cargo=None, left=None, right=None):
        self.cargo = cargo
        self.left = left
        self.right = right

def cantidad_hojas(btree: BinaryTree) -> int:
    """
    Devuelve la cantidad de hojas que tiene el árbol.
    """

    if btree == None:
        return 0

    if btree.left is None and btree.right is None:
        return 1

    return cantidad_hojas(btree.left) + cantidad_hojas(btree.right)


def numeros_entre(btree: BinaryTree, l: int, r: int) -> list[int]:
    """
    Devuelve una lista con los números del árbol que están entre 'l' y 'r' inclusive.
    """
    if btree == None:
        return []

    numeros = []

    if l <= btree.cargo <= r:
        numeros.append(btree.cargo)

    numeros.extend(numeros_entre(btree.left, l, r))
    numeros.extend(numeros_entre(btree.right, l, r))

    return numeros

# otra forma
def numeros_entre2(btree: BinaryTree, l: int, r: int) -> list[int]:
    """
    Devuelve una lista con los números del árbol que están entre 'l' y 'r' inclusive.
    """
    if btree == None:
        return []

    numeros = []
    if l <= btree.cargo <= r:
        numeros = [btree.cargo] # lista parcial con la raíz

    numeros_izq = numeros_entre(btree.left, l, r)   # lista parcial con el subárbol izquierdo
    numeros_der = numeros_entre(btree.right, l, r)  # lista parcial con el subárbol derecho
    print(numeros)
    print(numeros_izq)
    print(numeros_der)
    return numeros + numeros_izq + numeros_der  # concatenación de las listas
```

```python
class Objeto:
    def __init__(self, nombre: str, peso: int):
        self.nombre = nombre
        self.peso = peso

    def __str__(self):
        return self.nombre

    __repr__ = __str__

    def __gt__(self, other):
        return self.peso > other.peso


def es_solucion(mochila: list[Objeto]) -> bool:
    """
    Verifica que el peso total de los objetos sea igual a
    la capacidad de la mochila (W).
    """
    return sum(objeto.peso for objeto in mochila) == W


def elegir_candidato(candidatos: list[Objeto]) -> Objeto:
    """
    Selecciona el objeto de mayor peso en la lista de candidatos.
    """
    max_peso = -1
    max_objeto = None

    for objeto in candidatos:
        if objeto.peso > max_peso:
            max_peso = objeto.peso
            max_objeto = objeto

    return max_objeto


def es_factible(mochila: list[Objeto]) -> bool:
    """
    Comprueba que el volúmen total no supere la capacidad de la mochila (W).
    """
    return sum(objeto.peso for objeto in mochila) <= W


# lista de objetos para la mochila
candidatos = [Objeto("w1", 5),
              Objeto("w2", 9),
              Objeto("w3", 8),
              Objeto("w4", 12),
              Objeto("w5", 7),
              Objeto("w6", 12)]

W = 33 # capacidad de la mochila

# resolución:

eleccion_actual = []

while not es_solucion(eleccion_actual):
    x = elegir_candidato(candidatos)
    candidatos.remove(x)
    if es_factible(eleccion_actual + [x]):
        eleccion_actual.append(x)

print(eleccion_actual)
```

```python
# Encontrar el mayor elemento en L con D&C
def es_caso_base(problema: list[int]) -> bool:
    """
    Devuelve True si la lista tiene un solo elemento.
    """
    return len(problema) == 0 or len(problema) == 1

def resolver_caso_base(problema: list[int]) -> list[int]:
    """
    Devuelve el único elemento de la lista.
    """
    if len(problema) == 0:
        return None
    return problema[0]

def dividir(problema: list[int]) -> tuple[list[int], list[int]]:
    """
    Recibe una lista, la divide por la mitad y devuelve esas dos
    partes como una tupla de listas.
    """
    m = len(problema) // 2
    subp1 = problema[:m]
    subp2 = problema[m:]
    return subp1, subp2

def combinar(sol1: list[int], sol2: list[int]) -> int:
    """
    Recibe dos listas de un número cada una y
    devuelve el mayor de los dos.
    """
    return max(sol1, sol2)

def resolver(problema: list[int]) -> int:
    """
    Verifica que sea un caso base y lo resuelve, sino, divide la lista por la mitad y resuleve
    recursivamente los subproblemas hasta encontrar el mayor número de la lista.
    """
    if es_caso_base(problema):
        return resolver_caso_base(problema)

    subp1, subp2 = dividir(problema)
    sol1 = resolver(subp1)
    sol2 = resolver (subp2)
    return combinar(sol1, sol2)


L = [1, 2, 3, 4, 5, 1, 0, 3, 5, 1, 0, 10, 5]
print(resolver(L))
```

```python


class ConeVolumeIntegral(ThreeDScene):
    def construct(self):
        self.set_camera_orientation(phi=75 * DEGREES, theta=-45 * DEGREES)

        # Crear un cono
        cone = Cone(base_radius=1, height=2, direction=OUT)
        cone.set_fill(PINK, opacity=0.5)
        cone.set_stroke(PINK)

        # Añadir el cono a la escena
        self.add(cone)

        # Configurar las ecuaciones y el texto
        integral_tex = MathTex(
            r"V = \int_{0}^{h} \pi r^2 \, dz"
        ).scale(0.75).to_edge(UP)

        explanation = Text("Volumen de un cono usando una integral definida").scale(0.5).next_to(integral_tex, DOWN)

        self.add_fixed_in_frame_mobjects(integral_tex, explanation)

        # Animar la visualización de la integral definida
        self.play(Create(cone), Write(integral_tex), Write(explanation))
        self.wait(2)

        # Mostrar la altura h y el radio r
        height = Line(start=cone.get_bottom(), end=cone.get_top(), color=YELLOW)
        radius = Line(start=cone.get_bottom(), end=cone.get_bottom() + RIGHT, color=BLUE)

        h_label = MathTex("h").next_to(height, LEFT).scale(0.75)
        r_label = MathTex("r").next_to(radius, DOWN).scale(0.75)

        self.play(Create(height), Create(radius), Write(h_label), Write(r_label))
        self.wait(2)

        # Mostrar la superficie de integración
        surface = ParametricSurface(
            lambda u, v: np.array([
                (1 - u / 2) * np.cos(v),
                (1 - u / 2) * np.sin(v),
                u
            ]),
            u_range=[0, 2],
            v_range=[0, TAU],
            checkerboard_colors=[PURPLE_D, PURPLE_E],
            resolution=(15, 32)
        )

        self.play(Create(surface))
        self.wait(2)

        # Terminar la animación
        self.move_camera(phi=90 * DEGREES, theta=-90 * DEGREES, run_time=3)
        self.wait(2)

        # Texto final
        final_text = Text("El volumen del cono es V = 1/3 π r² h").scale(0.75).to_edge(DOWN)
        self.add_fixed_in_frame_mobjects(final_text)
        self.play(Write(final_text))
        self.wait(3)

if __name__ == "__main__":
    from manim import config
    config.media_width = "75%"
    scene = ConeVolumeIntegral()
    scene.render()
```
