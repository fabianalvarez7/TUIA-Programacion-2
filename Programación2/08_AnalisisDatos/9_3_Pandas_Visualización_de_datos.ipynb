# Visualización de datos

Un apartado fundamental de cualquier análisis o estudio de datos es la _visualización_ de los mismos.

Veremos diferentes formas de graficar los datos utilizando `matplotlib` y `pandas`.

Es fundamental que antes de realizar el análisis visual nos aseguremos de que los datos estan limpios y ordenados.

## GDP: Grafico de linea

Utilizaremos datos del crecimiento en el producto bruto interno (GDP por sus siglas en inglés).

```python
import pandas as pd
```

```python
gdp2 = pd.read_csv("gdp2.csv")
```

```python
gdp2.head()
```

Seteamos 'Country Code' como indice

```python
gdp2 = gdp2.set_index('Country Code')
```

```python
gdp2
```

### Analisis de datos faltantes

```python
gdp2.isna().sum()
```

En la columna 1960 nos faltan todos los datos, la removemos. Ademas, quitamos la fila 'Unnamed: 67' que no contiene valores.

```python
gdp2 = gdp2.dropna(axis=1, how='all')
gdp2.isna().sum()
```

Veamos cuantos datos nos faltan por índice.

```python
gdp2.isna().sum(axis=1)
```

Quitamos los paises donde nos falta algun valor

```python
gdp2 = gdp2.dropna(axis=0, how='any')
```

```python
gdp2.head()
```

### Guardar un dataframe en un archivo

```python
gdp2.to_csv('gdp_procesado.csv')
```

Para mostrar la evolución historica, usualmente utilizamos un grafico de linea. Veamos como podemos hacerlo con pandas, graficando la evolución del crecimiento del producto bruto interno para Argentina.

```python
arg = gdp2.loc['ARG'].copy()
arg
```

```python
# Quitamos todas las columnas que no sean numeros, que nos molestan para graficar.
# No nos preocupamos por el DF original, estamos trabajando sobre un copia!
# Ademas hay que convertir esos numeros convertidos en strings de nuevo a numero
# print(arg.index)
arg = arg.drop(index=['Country Name', 'Indicator Name', 'Indicator Code'])
arg = arg.apply(lambda x: float(x))
arg
```

```python
arg.plot.line()
```

Si lo necesitamos podemos agregar títulos, leyendas y nombres a los ejes utilizando `matplotlib`.

```python
import matplotlib.pyplot as plt

plt.title("Evolución del crecimiento del GDP en Argentina")
plt.ylabel("Crecimiento en puntos porcentuales")
arg.plot.line()
```

Podemos visualizar varias series de datos temporales en el mismo grafico para analizar la relación entre ambas. Tambien podemos graficar solo un subconjunto de los datos.

Veamos un ejemplo realizando un grafico que permita comparar la evolucion del GDP entre AFE y AFW durante la decada del 90.

```python
datos = gdp2.loc[['AFE', "AFW"], ['1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999']].copy()
datos
```

```python
datos.plot.line()
```

```python
datos.transpose().plot.line()
```

## NBA: Grafico de caja y bigote

Cuando estamos interesados en visualizar la _dispersión_ de una variable, es normal utilizar un grafico de caja y bigote, tambien llamado _boxplot_. En este tipo de grafico, vemos cuales son los valores tipicos de las variables y cuales son sus valores atipicos. Lo normal es utilizarla para observar la dispersión de una variable continua, respecto de una variable categorica.

Para visualizar esto utilizaremos un dataset de jugadores de la NBA.

```python
nba = pd.read_csv('NBAStats.csv')
nba
```

```python
nba.isnull().sum()
```

```python
nba = nba.dropna(axis=1)
nba.isnull().sum()
```

```python
nba.boxplot("AGE", by="POS")
```

```python
nba.boxplot("AGE", by="POS")

plt.title("Distribución de edad agrupando por posición")
plt.xlabel("Posicion")
plt.ylabel("Edad (Años)")
```

## Grafico de torta.

Sirve para ver porcentajes.

Se utilizan en aquellos casos donde interesa no mostrar el número de veces que ocurre una variable categorica, si no que interesa visualizar mejor la proporción en que aparece esa característica respecto del total.

A pesar de su popularidad, se trata de un tipo de gráfico poco recomendable debido a que nuestra capacidad perceptual para estimar relaciones de proporción o diferencias entre áreas de sectores circulares es mucho menor que, por ejemplo, entre longitudes o posiciones, tal y como sucede en otras gráficas.



```python
nba.groupby('POS').count()['NAME'].plot.pie()
```

## Grafico de barras

 Los gráficos de barras pueden ser usados para comparar cantidades de una variable categorica. Las barras pueden orientarse horizontal o verticalmente.

```python
nba.groupby('POS').count()['NAME'].plot.bar()
```
