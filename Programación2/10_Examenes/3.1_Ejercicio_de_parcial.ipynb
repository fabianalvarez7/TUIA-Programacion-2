**Ejercicio 4**: Utilizando la misma clase Tree de la sección anterior, implemente otra clase llamada BSTree que herede de esta, reimplemente los métodos **menor_mayor**, **buscar** e implemente un nuevo método llamado **insertar** que inserte un elemento.

````python
class BSTree(Tree):
  # Falta resolver menor_mayor

  def buscar(self, element):
    if self.cargo == element:
      return True
    
    elif self.cargo <= element:
      # Sigo buscando a la derecha - si hay!
      if self.right:
        return self.right.buscar(element)
      else:
        return False

    elif self.cargo >= element:
      # Sigo buscando a la izquierda - si hay!
      if self.left:
        return self.left.buscar(element)
      else:
        return False

  def insert(self, element):
    if self.cargo == element:
      return
    
    elif self.cargo <= element:
      # Sigo buscando a la derecha - si hay!
      if self.right:
        self.right.insert(element)
      else:
        self.right = BSTree(element)

    elif self.cargo >= element:
      # Sigo buscando a la izquierda - si hay!
      if self.left:
        self.left.insert(element)
      else:
        self.left = BSTree(element)

import random
btree = BSTree(10)
for i in range(10):
  btree.insert(random.randint(0, 20))

print(btree.menor_mayor())
print(btree.buscar(10))
print(btree.buscar(100))
print(btree.buscar(-10))
print(btree.buscar(5))
````
