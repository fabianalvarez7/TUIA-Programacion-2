```python
def funcion_recursiva(x):
    if x == 40:
        return x
    else:
        return x + funcion_recursiva(x*2)

funcion_recursiva(5)  
```

```python
def stri(s):
    x = len(s)
    if x > 0:
        t = s [0:x-1]
        stri(t)
        print(s)

stri("COMPSCO")
```

```python
def rec(x):
    if x == 1:
        return 1
    else:
        return rec(x-1) + rec(x-3)

rec(5)
```

```python
def mystry(x):
    if x == 0:
        return 1
    return mystry(x-1) + mystry(x-1)

mystry(5)
```
