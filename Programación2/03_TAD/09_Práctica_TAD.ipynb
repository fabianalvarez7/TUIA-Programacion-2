**Ejercicio 1**: Definir la interfaz, con la estructura que vimos en teoría, de un TAD para representar una celda. Una celda es capaz de contener un **único** valor a la vez, es decir, siempre que ponemos un nuevo valor, debemos quitar el valor anterior.

_Nota: no olvidar describir el \_\_init___

Métodos:

__init__: inicializa una celda vacía

tiene_algo: dice si tiene un valor o no

ver: dice el valor de la celda

asignar: dado un valor lo ubica dentro de la celda

sacar: si hay un valor lo saca y deja la celda vacía

**Ejercicio 2**: Basándose en la interfaz hecha en el ejercicio anterior, implementar en Python una clase Celda que respete el comportamiento descrito. Escribir código cliente de dicha interfaz.

````python
from typing import Any

class Celda:
    def __init__(self):
        self.valor = None

    def __str__(self):
        return str(self.valor)

    def tiene_algo(self):
        return self.valor is not None

    def ver(self) -> Any:
        return self.valor

    def asignar(self, valor: Any):
        self.valor = valor

    def sacar(self) -> None:
        self.valor = None


micelda = Celda()
micelda.asignar(5)
micelda.ver
print(micelda)
micelda.asignar(10)
print(micelda)
micelda.sacar
print(micelda)
````

````python
class Vector():
    def __init__(self, dimension):
        self.dimension = dimension
        self.numeros = [] * dimension

    def __str__(): #completar
        pass

    def get(self, pos)-> float | None:
        if 1 <= pos <= self.dimension:
            return self.numeros[pos - 1]
        return None

    def put(self, pos: int, numero: float):
        if 1 <= pos <= self.dimension:
            self.numeros[pos - 1] = numero
            return numero
        return None

    def __add__ (self, other: Vector) -> Vector:
        if self.dimension != other.dimension:
            return None

        resultado = Vector(self.dimension)

        for pos in range(1, self.dimension +1):
            suma = self.get + other.get(pos)

        resultado.put(pos, suma)
````

**Ejercicio 8**: Implementar el TAD Pila (Stack) utilizando listas de python

````python
from typing import Any

class Pila:
    def __init__ (self) -> None:
        self.items = []

    def push (self, item: Any) -> None:
        self.items.append(item)

    def pop (self) -> Any:
        if self.isEmpty():
            print("La pila está vacía")
            return
        return self.items.pop()

    def isEmpty(self) -> bool:
        return (self.items == [])

p = Pila()
p.isEmpty()
p.push(1)
p.isEmpty()
p.push(5)
p.push(22)
p.pop()
q = Pila()
q.pop()
````
