## Pila

**Ejercicio 1**: Implementar el TAD Pila (Stack) utilizando listas de python

````python
from typing import Any

class Pila:
    """Representa una pila con operaciones de apilar, desapilar y
    verificar si est치 vac칤a. LIFO"""

    def __init__(self) -> None :
        """Crea una pila vac칤a."""
        self.items = []


    def push(self, item: Any) -> None:
        """Apila un elemento."""
        self.items.append(item)


    def pop(self) -> Any:
        """Desapila un elemento y lo devuelve.
        Si la pila est치 vac칤a, imprime un mensaje de error
        y retorna inmediatamente."""
        if self.isEmpty():
            print("La pila est치 vac칤a")
            return
        return self.items.pop()


    def isEmpty(self) -> bool:
        """Devuelve True si la pila est치 vac칤a y False si no."""
        return (self.items == [])
````

**Ejercicio 2**: Implementar el mismo TAD, pero utilizando en la implementaci칩n una estructura enlazada.



````python
class _Nodo:
    def __init__(self, dato: Any = None, prox = None):
        self.dato = dato
        self.prox = prox


class PilaEnlazada:
    """Modela una pila enlazada."""

    def __init__(self) -> None:
        """Crea una pila vac칤a."""
        self.items = []
        self.prim = None
        self.len = 0


    def __str__(self):
        s = "["
        item = self.prim
        while item is not None:
            s += str(item.dato) + " <- "
            item = item.prox
        s += "]"
        return s


    def __len__(self):
        """Devuelve la longitud de la pila"""
        return self.len


    def push(self, item: Any) -> None:
        """Apila un elemento."""
        nuevo = _Nodo(item)
        # caso especial: la lista est치 vac칤a
        if self.len == 0:
            self.prim = nuevo   # el nuevo nodo es el primer nodo
        else:
            ultimo = self.prim
            while ultimo.prox is not None:  # se recorre la lista hasta el 칰ltimo nodo
                ultimo = ultimo.prox
            ultimo.prox = nuevo     # se actualiza el nodo
        self.len += 1


    def pop(self) -> Any:
        """Desapila el 칰ltimo elemento que fue agregado y lo devuelve.
        Si la pila est치 vac칤a, imprime un mensaje de error
        y retorna inmediatamente."""

        if self.isEmpty():
            print("La pila est치 vac칤a")
            return
        else:
            nodo_anterior = self.prim
            nodo_actual = nodo_anterior.prox
            for pos in range(1, self.len):
                nodo_anterior = nodo_actual
                nodo_actual = nodo_anterior.prox
            # Guardar el dato y descartar el nodo
            dato = nodo_actual.dato
            nodo_anterior.prox = nodo_actual.prox
            self.len -= 1
            return dato


    def isEmpty(self) -> bool:
        """Devuelve True si la pila est치 vac칤a y False si no."""
        return self.len == 0
````

````python
miPila = PilaEnlazada()
miPila.push("banana")
miPila.push("manzana")
miPila.push("pera")
miPila.push("naranja")
miPila.isEmpty()
print(miPila)
````

## Cola

**Ejercicio 5**: Bas치ndose en el TAD de cola, implementar una clase Queue con listas de Python.

````python
class Cola:
    """Representa una cola, con operaciones de encolar y desencolar.
    El primero en ser encolado es el primero en ser desencolado. FIFO"""

    def __init__(self) -> None:
        """Crea una cola vac칤a"""
        self.items = []


    def insert(self, x: Any) -> None:
        """Encola el elemento x"""
        self.items.append(x)


    def remove(self) -> Any:
        """Elimina el primer elemento de la cola y devuelve su valor.
        Si la cola est치 vac칤a, imprime un error y retorna"""
        if self.isEmpty():
            print("La cola est치 vac칤a")
            return
        return self.items.pop(0)


    def isEmpty(self) -> bool:
        """Devuelve True si la cola est치 vac칤a y False si no."""
        return len(self.items) == 0
````

### Cola enlazada

````python
class _Nodo:
    def __init__(self, dato: Any = None, prox = None):
        self.dato = dato
        self.prox = prox


class Queue:
    """Modela una cola enlazada."""

    def __init__(self) -> None:
        """Crea una cola enlazada vac칤a."""
        self.head = None
        self.length = 0


    def __str__(self):
        s = "[ <- "
        item = self.head
        while item is not None:
            s += str(item.dato) + " <- "
            item = item.prox
        s += "]"
        return s


    def insert(self, valor: Any) -> None:
        """Encola un elemento."""
        node = _Nodo(valor)
        node.prox = None
        # caso especial: la lista est치 vac칤a
        if self.head == None:
            self.head = node   # el nuevo nodo es el primer nodo
        else:   # se busca el 칰ltimo nodo de la lista
            last = self.head
            while last.prox is not None:  # se recorre la lista hasta el 칰ltimo nodo
                last = last.prox
            last.prox = node     # se agrega el nodo al final
        self.length += 1


    def remove(self) -> Any:
        """Desencola el primer elemento que fue agregado y lo devuelve.
        Si la cola est치 vac칤a, imprime un mensaje de error
        y retorna inmediatamente."""

        if self.isEmpty():
            print("La cola est치 vac칤a")
            return
        else:
            dato = self.head.dato
            self.head = self.head.prox
            self.length -= 1
            return dato


    def isEmpty(self) -> bool:
        """Devuelve True si la cola est치 vac칤a y False si no."""
        return self.length == 0
````

### Cola enlazada mejorada

````python
class _Nodo:
    def __init__(self, dato: Any = None, prox = None):
        self.dato = dato
        self.prox = prox


class ImprovedQueue:
    def __init__(self) -> None:
        self.length = 0
        self.head = None
        self.last = None


    def isEmpty(self) -> bool:
        return self.length == 0


    def insert(self, valor: Any) -> None:
        node = _Nodo(valor)
        node.prox = None
        if self.length == 0:
            # si la lista est치 vac칤a, el nuevo nodo es el primero y el 칰ltimo
            self.head = self.last = node
        else:
            # localizar el 칰ltimo nodo
            last = self.last
            # agregar el nuevo nodo al final
            last.prox = node
            self.last = node
        self.length += 1


    def remove(self) -> Any:
        dato = self.head.dato
        self.head = self.head.prox
        self.length -= 1
        if self.length == 0:
            self.last = None
        return dato
````
