TAD Cola de prioridad (Priority Queue)

```python
class PriorityQueue:
    def __init__(self):
        self.items = []

    def isEmpty(self):
        return self.items == []

    def insert(self, item):
        self.items.append(item)

#política de prioridad: eliminar siempre el elemento mayor

    def remove(self):
        maxi = 0
        for i in range(1, len(self.items)):
            if self.items[i] > self.items[maxi]:
                maxi = i
        item = self.items[maxi]
        self.items[maxi:maxi + 1] = []
        return item
```

```python
#probamos la implementación:
q = PriorityQueue()
q.insert(11)
q.insert(12)
q.insert(14)
q.insert(13)

while not q.isEmpty():
    print (q.remove())
```

Estructuras de datos enlazadas. Clase Nodo (Node)

```python
class Nodo:
    def __init__(self, cargo=None, next=None):
        self.cargo = cargo
        self.next = next

    def __str__(self):
        return str(self.cargo)
```

Cola enlazada

```python
class Queue:
    def __init__(self):
        self.length = 0
        self.head = None

    def isEmpty(self):
        return (self.length == 0)

    def insert(self, cargo):
        node = Nodo(cargo)
        node.next = None

        if self.head == None:
            #si la lista está vacía, el nuevo nodo se ubica al principio
            self.head = node
        else:
            #localizar el último nodo de la lista
            last = self.head
            while last.next:
                last = last.next
            #agragar al final el nuevo nodo
            last.next = node
        self.length = self.length + 1

    def remove(self):
        cargo = self.head.cargo
        self.head = self.head.next
        self.length = self.length - 1
        return cargo
```

Ejercicio 1: Definir la interfaz de un tipo abstracto de datos para representar una celda.
Una celda es un TAD capaz de contener un único valor a la vez, es decir, siempre que
ponemos un nuevo valor, debemos quitar el valor anterior. Usar lenguaje natural.

```python
class Celda:
    def __init__(self):
        self.valor = None

    def __str__(self):
        return str(self.valor)

    def insertar(self, valor):
        if self.valor is not None:
            del self.valor
        self.valor = valor

```

Ejercicio 3: Definir la interfaz de un Conjunto Inmutable. Los conjuntos inmutables son
típicos en matemática. Se definen con elementos preestablecidos, nunca cambian, y la unica
operacion posible es preguntarnos si un elemento pertenece o no al conjunto.

```python
class Conjunto:
    def __init__(self, *elementos):
        self.elementos = set(elementos)

    def pertenece(self, elemento):
        return elemento in self.elementos
```
