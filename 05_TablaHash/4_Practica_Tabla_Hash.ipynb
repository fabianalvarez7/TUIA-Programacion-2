**Ejercicio 1**: Se quiere rellenar el stock de un supermercado. Se sabe que los códigos de los productos van del 000 al 999. Se ingresa alguna cantidad arbitraria de códigos (no necesariamente todos) junto con la cantidad que hay en stock. Implemente una tabla de direccionamiento directo, implemente los métodos `insert` y `lookup` y utilícela para resolver.

```python
class DirectAccessTable():
    def __init__(self, capacity):
        self.m = capacity
        self.T = [None] * self.m

    def insert (self, key, element):
        self.T[key] = element

    def search (self, key):
        return self.T[key]

    def delete (self, key):
        self.T[key] = None
```

```python
stock = DirectAccessTable (10)
stock.insert(0, 10)
stock.insert(1, 20)
stock.insert(2, 30)
stock.insert(3, 40)
#stock.insert(4, 50)
stock.insert(5, 60)
stock.insert(6, 70)
stock.insert(7, 80)
#stock.insert(8, None)
stock.insert(9, 100)

print(stock.T)
```

```python
print(stock.search(3))
print(stock.search(4))
```

```python
stock.delete(9)
print(stock.search(9))
print(stock.T)
```

**Ejercicio 4**: La siguiente función implementa una función de hash para strings conocida como el método de Horner. Explique porque es una buena función de hash.

```python
def horner(string, M=487): # M es el tamaño de la tabla - deberia ser primo!
  A = 37 # Este valor se puede cambiar, algunos valores elegidos usualmente son 37 y 256
  result = 0
  for character in string:
    result = ord(character) + A * result
  return result % M

print(horner("hola"))
print(horner("halo"))
print(horner("hello") != horner("halo"))
```
